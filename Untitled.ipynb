```python
import numpy as np
import scipy.ndimage
import scipy.signal
import matplotlib.pyplot as plt
plt.style.use('dark_background')
import pepper
import imageio
```

```python
img = imageio.imread("Charge13+_Volt166_Mass48_Brightness0_Gain0_Shutter2722_Trigger_Off_T_Delay_us40_PulseLength_us400_01-02-2017_15h30m57_OriginalPicture.png")
```

```python
img = img.copy()
```

```python
def show_img(img):
    plt.figure(figsize=(15,10))
    plt.imshow(img)
    plt.colorbar()
    plt.show()
```

```python
show_img(img)
```

```python
def determine_background(img):
    return img[:100, :100].mean()
```

```python
determine_background(img)
```

```python
img2 = img - determine_background(img)
show_img(img2)
```

```python
img3 = scipy.ndimage.median_filter(img2, size=2)
show_img(img3)
```

```python
img4 = np.clip(img3, 1 , None)
show_img(np.log(img4))
```

```python
plt.figure()
plt.plot(img4.mean(0))

plt.figure()
plt.plot(img4.mean(1))

plt.show()
```

```python
peaks, _ = scipy.signal.find_peaks(img4.mean(0), distance=30, height=60)
plt.figure(figsize=(8,8))
plt.plot(img4.mean(0))
plt.semilogy(peaks, img4.mean(0)[peaks],"o")
seps = (peaks[:-1] + peaks[1:])/2
seps = np.insert(seps, 0, 2*seps[0]-seps[1])
seps = np.append(seps, 2*seps[-1]-seps[-2])
plt.vlines(seps, 0 ,800, "w", linewidth=.5)
```
